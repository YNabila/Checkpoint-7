## 1.Print the three first rows using the head() method.

```python
import numpy as np
import pandas as pd
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
exam_df=pd.DataFrame(exam_data,index=labels)
```

```python
exam_df.head(n=3)
```

```python
exam_df
```

## 4.Write a Pandas program to append a new row 'k' to the DataFrame with these values (name: "Suresh", score: 15.5, attempts: 1, qualify: "yes").

```python
exam2={'name': "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
```

```python
label=['k']
new_row=pd.DataFrame(exam2,index=label)
```

```python
exam_df=pd.concat([exam_df,new_row])
```

```python
exam_df
```

## 5.Write a Pandas program to delete the 'attempts' column from the DataFrame.

```python
exam_df.drop("attempts", axis=1)
```

## 6.Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.

```python
new=[]
s=exam_df['score'].to_dict()
for key in s.keys():
    if s[key]>10:
        new.append(1)
    else:
        new.append(0)
new
```

```python
exam_df['success']=new 
```

```python
exam_df
```
